```python
# LSTM for sequence classification in the IMDB dataset
import numpy as np
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Bidirectional
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
import keras as kr
import random
```

## Map characters

```python
char_to_int = {}
int_to_char = []

count = 0

# Small letters
for i in range(26):
    char = chr(ord("a")+i)
    char_to_int[char] = count
    int_to_char.append(char)
    count += 1
    
# Capital letters
for i in range(26):
    char = chr(ord("A")+i)
    char_to_int[char] = count
    int_to_char.append(char)
    count += 1
    
# Decimal values
for i in range(10):
    char = chr(ord("0")+i)
    char_to_int[char] = count
    int_to_char.append(char)
    count += 1

char = "."
char_to_int[char] = count
int_to_char.append(char)
count += 1

char = "-"
char_to_int[char] = count
int_to_char.append(char)
count += 1

char = " "
char_to_int[char] = count
int_to_char.append(char)
count += 1

print("char_to_int: ", char_to_int)
#print(charDict["a"])
print("int_to_char: ",int_to_char)
```

## Generate data

```python
def add_spaces(cur_string,cur_x,cur_y,min_val,max_val):
    num_spaces = min_val + int(random.random()*max_val)

    for j in range(num_spaces):
        char = " "
        cur_string  += char
        cur_x.append(char_to_int[char])
        cur_y.append(0)
    return cur_string,cur_x,cur_y

def add_date(cur_string,cur_x,cur_y):
    month = 1+int(random.random()*12.0)
    day = 1+int(random.random()*30.0)
    
    date_string = str(month) + "-" + str(day)
    
    for j in range(len(date_string)):
        char = date_string[j]
        cur_string += char
        cur_x.append(char_to_int[char])
        cur_y.append(1)
    return cur_string,cur_x,cur_y

def random_discription(cur_string,cur_x,cur_y,disc_len):
    for j in range(disc_len):
        rand_int = int(random.random()*26)
        char = chr(ord("a")+rand_int)
        cur_string  += char
        cur_x.append(char_to_int[char])
        cur_y.append(2)
    
    return cur_string,cur_x,cur_y

def add_rand_value(cur_string,category,cur_x,cur_y):
    
    char = "R"
    cur_string  += char
    cur_x.append(char_to_int[char])
    cur_y.append(category)
    
    rand_float = round(random.random()*1000.0,2)
    
    value_as_string = str(rand_float)
    
    for j in range(len(value_as_string)):
        char = value_as_string[j]
        cur_string  += char
        cur_x.append(char_to_int[char])
        cur_y.append(category)
    
    return cur_string,cur_x,cur_y

def generate_text_data(num_data=40):
    x_data = []
    x_text = []
    
    y_data = []
    
    for i in range(num_data):
        
        cur_string = ""
        cur_x = []
        cur_y = []
        
        # Add a random amount of spaces
        cur_string,cur_x,cur_y = add_spaces(cur_string,cur_x,cur_y,0,2)
        
        # Add a random date
        cur_string,cur_x,cur_y = add_date(cur_string,cur_x,cur_y)
        
        # Add random amount of spaces
        cur_string,cur_x,cur_y = add_spaces(cur_string,cur_x,cur_y,1,2)
        
        # Add a random description
        cur_string,cur_x,cur_y = random_discription(cur_string,cur_x,cur_y,20)
        
        # Add random amount of spaces
        cur_string,cur_x,cur_y = add_spaces(cur_string,cur_x,cur_y,0,2)
        
        # Add a random rand value
        cur_string,cur_x,cur_y = add_rand_value(cur_string,3,cur_x,cur_y)
        
        # Add random amount of spaces
        cur_string,cur_x,cur_y = add_spaces(cur_string,cur_x,cur_y,0,2)
        
        # Add a random rand value
        cur_string,cur_x,cur_y = add_rand_value(cur_string,4,cur_x,cur_y)
        
        x_text.append(cur_string)
        x_data.append(cur_x)
        y_data.append(cur_y)
        
    x_data = np.array(x_data)
    y_data = np.array(y_data)

    return x_data, y_data, x_text
```

```python
def zero_pad_to_numpy(data,catagory_size,max_len= None):
    if max_len==None:
        max_len = 0

        for j in range(len(data)):
            if len(data[j])>max_len:
                max_len = len(data[j])
    
    new_data = np.zeros((len(data),max_len,catagory_size),dtype=np.int)
    
    for j in range(len(data)):
        for k in range(len(data[j])):
            new_data[j][k][data[j][k]] = 1
            
        for k in range(len(data),max_len-len(data[j])):
            new_data[j][k][0] = 1
    return new_data
```

```python
np.random.seed(1330)

# Hyper parameters
num_data = 10000

# Generate the data
x_train,y_train,x_train_text = generate_text_data(num_data)

print("Text: ", x_train_text[1000])
print("Int input: ", x_train[1000])
print("Category output: ", y_train[1000])

input_cat = len(int_to_char)
output_cat = 5

# Zero pad the data and convert the data into the correct format
x_train = zero_pad_to_numpy(x_train,input_cat)
y_train = zero_pad_to_numpy(y_train,output_cat)
print("Category output numpy: ", y_train[1000])
```

```python
def createModel(input_dim,output_cat):
    model=Sequential()
    lstmPar = LSTM(input_dim = input_cat, output_dim = 50, return_sequences = True)
    model.add(Bidirectional(lstmPar, input_shape=(None, input_cat)))
    model.add((Dense(output_dim = 50)))
    model.add((Dense(output_dim = output_cat, activation  =  "softmax")))
    model.compile(loss = "mse",optimizer='adam', metrics=['accuracy'])
    #model.compile(loss = "mse", optimizer = "rmsprop")
    return model
```

```python
# Create and fit the model
input_cat = len(int_to_char)
output_cat = 5
model = createModel(input_cat,output_cat)
model.fit(x_train, y_train, nb_epoch = 10, batch_size = 400)
```
