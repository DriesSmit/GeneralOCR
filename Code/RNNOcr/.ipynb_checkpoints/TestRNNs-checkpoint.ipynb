```python
# LSTM for sequence classification in the IMDB dataset
import numpy as np
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
import keras as kr
```

# Static example

```python
np.random.seed(1337)

sample_size = 256
x_seed = [1, 0, 0, 0, 0, 0]
y_seed = [1, 0.8, 0.6, 0, 0, 0]

x_train = np.array([[x_seed] * sample_size]).reshape(sample_size,len(x_seed),1)
y_train = np.array([[y_seed]*sample_size]).reshape(sample_size,len(y_seed),1)

print(x_train.shape)

model=Sequential()
model.add(LSTM(input_dim  =  1, output_dim = 50, return_sequences = True))
model.add((Dense(output_dim = 1, activation  =  "sigmoid")))
model.compile(loss = "mse", optimizer = "rmsprop")
model.fit(x_train, y_train, nb_epoch = 100, batch_size = 32)

print(model.predict(np.array([[[1],[0],[0],[0],[0],[0]]])))
```

# Copy previous input

```python
def createModel():
    model=Sequential()
    model.add(LSTM(input_dim  =  1, output_dim = 50, return_sequences = True))
    model.add((Dense(output_dim = 50, activation  =  "sigmoid")))
    model.add((Dense(output_dim = 1)))
    model.compile(loss = "mse", optimizer = "rmsprop")
    return model
```

```python
np.random.seed(1337)

in_size = 5000
seq_lenth = 10

x_train = np.random.rand(in_size,seq_lenth)

y_train = x_train[:,0:-1]
y_train = np.insert(y_train, 0, 0.0, axis=1)

x_train = x_train.reshape((in_size,seq_lenth,1))
y_train = y_train.reshape((in_size,seq_lenth,1))

#print(x_train)
#print(y_train)

model = createModel()
model.fit(x_train, y_train, nb_epoch = 20, batch_size = 32)

print(model.predict(np.array([[[1],[0.2],[1],[0],[0],[0]]])))
```

```python
def generateData(num_data=40,numCatagories=3,max_min_len=(5,30)):
    x_data = []
    y_data = []

    for i in range(num_data):
        num_samples = np.random.randint(max_min_len[0],max_min_len[1],1)
        X = np.zeros(num_samples,dtype=np.int)
        Y = np.zeros((num_samples,numCatagories),dtype=np.int)

        rand_indexes = np.random.randint(1,num_samples-1,numCatagories-1)

        for rand_index in rand_indexes:
            X[rand_index] = 1

        count = 1
        for j in range(len(X)):
            if j+1<len(X) and X[j] == 1:
                count +=1
            Y[j][count-1] = 1

        x_data.append(X)
        y_data.append(Y)
    return x_data, y_data
```

```python
# Hyper parameters
num_data = 1
numCatagories = 2
max_min_len = 29,30

X_train,Y_train = generateData(num_data,numCatagories,max_min_len)
X_test,Y_test = generateData(num_data,numCatagories,max_min_len)

print(X_train)
print(Y_train)
```
